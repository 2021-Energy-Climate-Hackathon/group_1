# Analyze NAO influence on hourly temperature prediction

```python
# read in UK t2m & NAO data
df = pd.read_csv("country_data/ERA5_hourly_UK.csv")
df.columns = ["date_time","t2m","ssrd","msl","nao", "w10","w100"]
df.head()
```

```python
# add weeks/years columns
df["year"] = pd.to_numeric(pd.to_datetime(df.date_time).dt.strftime("%Y"))
df["week"] = pd.to_numeric(pd.to_datetime(df.date_time).dt.strftime("%U"))
df.head()
```

### Transform data into independent/dependent variables

```python
# average t2m/nao by day, take the max of week/year
df.date_time = pd.to_datetime(df.date_time)
df_daily = df.groupby(df.date_time.dt.date).mean()
df_daily.head()
```

```python
X = df_daily[["t2m","ssrd","w100","w10","msl","nao","week","year"]]
```

```python
# create 24-hour arrays as target variable
Y_list = []
for i in range(int(len(df)/24)):
    Y_list.append(df.t2m.iloc[(i*24):(i*24+24)].values)
Y = np.array(Y_list)

# add date_time index
Y = pd.DataFrame(Y,index=df_daily.index)
```

```python
# variable shapes
print(X.shape)
print(Y.shape)
```

```python
# randomly sample data 
x = X.loc[X['year'].isin([1980, 1985, 1990, 1995, 2000,2005,2010,2015,2020])]
```

```python
sns_plot = sns.pairplot(x, hue = 'year' , corner = True)
```

### Regression : t2m

Instead of treating the target variable as a vector containing hourly temperatures, convert hourly values to daily averages and standard deviations. This will provide insight into how the diurnal cycle varies in amplitude and overall temperature by NAO value. In addition, regression can be performed on single values.

```python
# create second dataframe containing metrics on the hourly values
means = Y.mean(axis=1)
std = Y.std(axis=1)

# Y2 contains average temp and std
Y2 = pd.DataFrame({"mean_t2m":means,"std_t2m":std})
Y2.index = Y.index
```

```python
# concat x and Y together 
xy = pd.concat([x,Y2],axis=1)

# create the linear regression using a univariate regression (NAO only regressor)
lr = smf.ols('std_t2m ~ nao',data=xy).fit()

print(lr.summary())
```

Overall NAO is a very poor predictor of average temperature. 

Very poor R-square.

F-statistic is significant although a low number.

NAO is statistically significant, although not very helpful given the over poor performance of the regression.

NAO coefficient is negative, suggests that NAO -ve results in warmer temperatures in the UK which is incorrect.

Kurtosis/DW suggest model assumptions likely violated.

```python
sns.regplot(x=xy.nao,y=xy.mean_t2m)
sns.set(rc={'figure.figsize':(15,8)})
plt.title("Comparisons of Average Daily Temperatures and the NAO")
```

Plot of average daily temperature vs NAO indicates a weak relationship, sloping negatively.

This is contrary to the NAO, should be sloping positive.

Try filtering by winter weeks as NAO is of higher importance in the winter.

```python
# filter by winter weeks
winterxy = xy[(xy.week>44) | (xy.week<9)]
```

```python
# plot of filtered weeks
sns.regplot(x=winterxy.nao,y=winterxy.mean_t2m)
sns.set(rc={'figure.figsize':(15,8)})
plt.title("Comparisons of Average Daily Temperatures and the NAO: WINTER WEEKS")
```

Filtering by the winter months visually indicates NAO -ve supporting colder regimes and NAO +ve supporting warmer regimes, which is in coalition with literature. 

Suggests that the week of the year is a major factor for the NAO. Therefore retry the regression using an interaction term of NAO/week and use the filtered data as input.

```python
# create the linear regression using a univariate regression (NAO only regressor)
lr2 = smf.ols('mean_t2m ~ nao+week',data=winterxy).fit()

print(lr2.summary())
```

Including week as an interaction improved the regression.

AIC, BIC decreased, whilst R-square increased. 

All terms are significant, and the interaction term coefficient is slightly positive.

Again model assumptions are likely violated.

Overall, NAO is slightly significant in the winter months. This index would be good to capture the synoptic picture and feed into an ML algorithm when downscaling the temporal resolution of the climatological data. 
Remains to be seen if NAO is signficant for SSRD and wind speeds. Imagining that it is indeed more signficant.
